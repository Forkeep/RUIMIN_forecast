```python
# 获取主力合约期货价格数据
# 查询铝对应的代码
# df = pro.fut_basic(exchange='SHFE', fut_type='1', fields='ts_code,symbol,name,list_date,delist_date')
main_heyue_map = pro.fut_mapping(ts_code='AL.SHF',start_date='20150101')
heyue_list = list(main_heyue_map['mapping_ts_code'].unique())
res = pro.fut_daily(ts_code=heyue_list[0],fields='ts_code,trade_date,close,settle,vol,amount,oi')
for ts_code in heyue_list:
    tmp = pro.fut_daily(ts_code=ts_code,fields='ts_code,trade_date,close,settle,vol,amount,oi')
    res = res.append(tmp) 

main_heyue_price_daily = pd.merge(main_heyue_map,res,how='inner',left_on=['trade_date','mapping_ts_code'],right_on=['trade_date','ts_code'])
main_heyue_price_daily = main_heyue_price_daily[['trade_date','ts_code_y','close','settle','vol','amount','oi']]
col=['trade_date','ts_code','close','settle','vol','amount','oi']
main_heyue_price_daily.columns = col
main_heyue_price_daily.to_csv('futures.csv',index=False)
```

## 数据格式对齐过程
- PPI -> CDP  ->  M2  -> CPI

```python
# 交易日历
# 结果：所有交易日的年月日数据

today = int(''.join(str(date.today()).split('-')))
calendar = pro.trade_cal(exchange='SHFE', start_date='20150101',end_date=today)
calendar['cal_month'] = calendar['cal_date'].apply(lambda x : int(str(x)[:-2]))
calendar = calendar[calendar['is_open'] == 1]
calendar = calendar[['cal_date','cal_month']]
# calendar
```

```python
#  年季月 展开到  日 维度变换

#  PPI
# ppi_yoy   PPI：全部工业品：当月同比
# ppi_mom   PPI：全部工业品：环比
# ppi_accu  PPI：生产资料：累计同比
ppi = pd.read_csv('PPI.csv')
ppi = ppi[['month','ppi_yoy','ppi_mom','ppi_accu']]
ppi = ppi.dropna()
ppi = ppi[ppi.month >= 201501]


# CPI
cpi = pd.read_csv('CPI.csv')
cpi = cpi[['month','nt_val','town_val','cnt_val']]
cpi = cpi[cpi['month']>=201501]


# M2
m2 = pd.read_csv('M2.csv')
m2 = m2[['month','m0','m1','m2']]
m2 = m2[m2['month'] >= 201501]


# GDP

def quarter2month(quarter):
    year = quarter[0:4]
    last = quarter[4:]
    month = ''
    if(last == 'Q1'):
        month = year+'01,'+year+'02,'+year+'03'
    elif(last == 'Q2'):
        month = year+'04,'+year+'05,'+year+'06'
    elif(last == 'Q3'):
        month = year+'07,'+year+'08,'+year+'09'
    elif(last == 'Q4'):
        month = year+'10,'+year+'11,'+year+'12'
    return month
gdp = pd.read_csv('GDP.csv')
gdp = gdp[['quarter','gdp','pi','si','ti']]
gdp = gdp[gdp['quarter']>='2015Q1']
# gdp['month_tmp'] = gdp['quarter'].apply(lambda x : x[4:])
gdp['month'] = gdp['quarter'].apply(quarter2month)
gdp['month'] = gdp['month'].map(lambda x:x.split(','))
gdp = gdp.explode('month')
gdp['month'] = gdp['month'].apply(lambda x : int(x))
```

```python
futures = pd.read_csv('futures.csv')
futures['trade_date'] = futures['trade_date'].apply(lambda x:int(x))
futures
```

```python
ppi_daily = pd.merge(calendar,ppi,how='left',left_on='cal_month',right_on='month')
cpi_daily = pd.merge(ppi_daily,cpi,how='left',left_on='month',right_on='month')
m2_daily = pd.merge(cpi_daily,m2,how='left',left_on='month',right_on='month')
gdp_daily = pd.merge(m2_daily,gdp,how='left',left_on='month',right_on='month')
# gdp_daily.describe()
# gdp_daily = gdp_daily.dropna(axis=0,subset=['month'])
del gdp_daily['month']
del gdp_daily['cal_month']
del gdp_daily['quarter']
gdp_daily['cal_date'] = gdp_daily['cal_date'].apply(lambda x:int(x))
all_daily = pd.merge(gdp_daily,futures,how='left',left_on='cal_date',right_on='trade_date')
all_daily.fillna(method = 'ffill',inplace=True)
all_daily.to_csv('all_features.csv',index=False)
```
